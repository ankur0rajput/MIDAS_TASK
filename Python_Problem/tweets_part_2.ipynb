```python
#Importing the libraries
import json
import texttable as tt
```

loading the json file

```python
with open('total_tweets.json', 'r') as f:
    data=json.load(f)
```

Initializing empty lists to collect information of the tweets

```python
tweet_id=[]    
text=[]
date_and_time=[]
likes=[]
retweet_count=[]
images_count=[]
```

total 296 tweets are extracted.

Extraction of id,text,created_at,favorite_count,retweet_count is easy because we can get these by using key only.

Main difficulty arises during extraction of number of images in a tweet

I am using try and except block beacuse It is not necessary that each of the tweet will contain images. If error occur while extracting number of images (It means no any media section is there) then append None in the list 'images_count'.

Media section of extended_entities can hold maximum of four images information. so we can not get any information about the further images used. However This condition not arises in our case because none of our tweets contain more than two images. 

```python
for i in range(296):
    #collecting Id of the tweets to a list named as tweet_id
    tweet_id.append(data[str(i)]["id"])
    
    #collecting text of the tweets to a list named as text
    text.append(data[str(i)]["text"])
    
    #collecting date and time of the tweets to a list named as date_and_time
    date_and_time.append(data[str(i)]["created_at"])
    
    #collecting the number of likes of the tweets to a list named as likes
    likes.append(data[str(i)]["favorite_count"])
    
    #collecting the number of retweets to a list named as retweet_count
    retweet_count.append(data[str(i)]["retweet_count"])
    
    #collecting number of images in the the tweets to list named as images_count
    try:
        if len((data[str(i)]["extended_entities"])["media"])>0:
            images_count.append(len((data[str(i)]["extended_entities"])["media"]))
    except:
        images_count.append(None)
        continue
```

I am using texttable to show the information in the tabular form. we may use some other methods also.

```python
tab = tt.Texttable()
headings = ['tweet_id','text','date_and_time','likes','retweet_count','images_count']
tab.header(headings)
```

```python
for row in zip(tweet_id,text,date_and_time,likes,retweet_count,images_count):
    tab.add_row(row)
```

```python
s = tab.draw()
```
